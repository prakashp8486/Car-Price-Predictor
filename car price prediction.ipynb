```python
x=data.drop('Price',axis=1)
```

```python
y=data.Price
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=.2)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
```

```python
ohe=OneHotEncoder()
```

```python
ohe.fit(x[['name','company','fuel_type']])
```

```python
ohe.categories_
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
pipe.score(x_test,y_test)
```

```python
# loop to get random x_train ,x_test,y_train,y_test
liss=[]
for index in range(1000):
    x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=.2)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(x_train,y_train)
    liss.append(pipe.score(x_test,y_test)*100)
    print(pipe.score(x_test,y_test)*100,index)
```

```python
liss[np.argmax(liss)]
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=.2,random_state=np.argmax(liss))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(x_train,y_train)
pipe.predict(x_test)
print(pipe.score(x_test,y_test)*100)
```
